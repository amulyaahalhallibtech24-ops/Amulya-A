```python
import requests
import pandas as pd


url = "https://remoteok.com/api"


response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})
data = response.json()


jobs = data[1:]


job_list = []
for job in jobs:
    job_list.append({
        "Company Name": job.get("company", "N/A"),
        "Job Role": job.get("position", "N/A"),
        "Location": job.get("location", "Remote"),
        "Tags": ", ".join(job.get("tags", [])) if job.get("tags") else "N/A"
    })


df = pd.DataFrame(job_list)


df.to_csv("remoteok_jobs.csv", index=False, encoding="utf-8")


print(f"✅ Data collected successfully! Total jobs: {len(df)}")
print("Saved as remoteok_jobs.csv")
df.head()  
```
